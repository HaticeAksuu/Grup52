Bu projede kullanılan veri seti, toplam 9115 adet cilt görüntüsünden oluşmakta olup, yedi farklı cilt hastalığı sınıfını içermektedir.

__mel (Melanom)__ sınıfında 1113 görsel yer almaktadır. Bu kötü huylu tümör, erken evrede tespit edilirse cerrahi müdahale ile tedavi edilebilir.

__nv (Benign Nevüs)__ sınıfı 6705 görsel ile en kalabalık sınıftır. Melanositlerden oluşan bu lezyonlar genellikle zararsız benlerdir ve çeşitli görsel varyantları bulunur.

__bkl (İyi Huylu Keratoz)__ sınıfında 1099 görsel bulunmaktadır. Bu sınıf, seboreik keratoz, solar lentigo ve liken planus benzeri keratozları kapsar.

__df (Dermatofibrom)__ sınıfında ise 115 görsel yer almaktadır. Bu, ciltte küçük travmalara bağlı gelişen, kahverengi tonlarda ve genellikle zararsız bir yapıdır.

__akiec (Aktinik Keratoz ve İntraepitelyal Karsinom)__ sınıfı 327 görselle temsil edilmektedir. Güneş ışığına maruz kalma sonucu oluşan ve ilerleyerek cilt kanserine dönüşebilen lezyonlardır.

__bcc (Bazal Hücreli Karsinom)__ sınıfı 514 görsel içerir. Genellikle metastaz yapmayan ancak tedavi edilmezse çevre dokulara zarar verebilen bir kanser türüdür.

__vasc (Vasküler Lezyonlar)__ sınıfı 142 görsel içerir ve anjiyom, damar tümörleri ve kanamalı cilt yapıları gibi oluşumları kapsamaktadır.
